# Bar Plot in Python

Bar plots are a common way to visualize categorical data by showing the values of each category as bars.

# When to Use a Bar Plot


Your x-axis (or y-axis) is categorical (e.g., names, groups, gender, etc.).

You want to compare quantities across different groups

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('employee_data.csv')
display(df.head())
```

```python
# count gender values

gc = df['gender'].value_counts()
display(gc)
```

```python
display(gc.index)
```

```python
display(gc.values)
```

```python
# bar plot in matplotlib

plt.figure(figsize=(4, 3))
plt.bar(gc.index, gc.values, color= 'g')

plt.xlabel('Gender', fontsize=14)
plt.ylabel('Count')
plt.title('Gender Distribution')

plt.xticks(fontsize=12, rotation=90)
plt.yticks(fontsize=12)


plt.savefig('bar_plot.png')
plt.show()
```

```python
# Bar plot seaborn

plt.figure(figsize=(4, 3))
sns.barplot(x=gc.index, y=gc.values, color='g')

plt.xlabel('Gender', fontsize=14)
plt.ylabel('Count')
plt.title('Gender Distribution')

plt.xticks(fontsize=12, rotation=90)
plt.yticks(fontsize=12)
plt.show()
```
