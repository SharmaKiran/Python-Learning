# **countplot → Counts Categories Automatically**
Used for: Counting the frequency of each category

Input: Only a categorical column

No need to pre-calculate anything:

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv('employee_data.csv')
display(df.head(2))
```

```python
plt.figure(figsize=(5, 3))
sns.set_theme(style='whitegrid')
sns.countplot(x='department', data=df)

plt.ylabel('Dept Count')

plt.show()
```

# barplot → Plots Summary Statistics
Used for: Showing the mean (or Summary statistic (mean, median, etc.)) of a numerical value grouped by categories

Input: One categorical column + one numeric column

You can choose the statistic with estimator=

```python
# Department wise average salary of each person
# department and other salary

dept_salary = df.groupby('department')['salary'].mean()
display(dept_salary)
```

```python
plt.figure(figsize=(5, 3))
sns.set_theme(style='whitegrid')
sns.barplot(x=dept_salary.index, y=dept_salary.values)

plt.ylabel('Average Salary')

plt.show()
```

```python
# Department wise median salary of each person
# department and other salary

dept_salary_M = df.groupby('department')['salary'].median()
display(dept_salary_M)
```

```python

plt.figure(figsize=(5, 3))
sns.set_theme(style='whitegrid')
sns.barplot(x=dept_salary_M.index, y=dept_salary_M.values)

plt.ylabel('Median Salary')

plt.savefig('Medianj_Salary.png')

plt.show()
```
