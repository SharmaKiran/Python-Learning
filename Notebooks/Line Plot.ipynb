# **Line Plot**

A line plot is a type of chart used to display information as a series of data points ("markers") connected by straight lines. In Python, it's commonly used to visualize trends over time or ordered data.

# **When to Use a Line Plot**

* When you want to show changes over time (e.g., stock prices, temperature).

* To show trends or patterns in data.

* For continuous datasets.

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

import datetime
```

```python
df = pd.read_csv('employee_data.csv')
display(df.head(2))
```

```python
# ensure dat in datetime format

df['joining date'] = pd.to_datetime(df['joining date'])
display(df.head(2))
```

```python
d = pd.to_datetime(df['joining date'])
```

```python
# extract year and month nad save back to data

df['year']= df['joining date'].dt.year
display(df.head(2))
```

```python
df['month']= df['joining date'].dt.month
display(df.head(2))
```

```python
# plot line as year wise count of employess

count = df['year'].value_counts().sort_index()
count
```

```python
# Matplotlib

plt.figure(figsize=(6,2))
plt.plot(count.index, count.values, linestyle ='--', marker='o', color='red')
plt.xlabel('Year', color='g')
plt.ylabel('Count')
plt.title('Year Wise Employee Count')
plt.xticks(rotation=90, fontsize=14,color='b')

plt.show()
```

```python
# seaborn
plt.figure(figsize=(6,2))
sns.lineplot(x=count.index, y=count.values,
             linestyle ='--', marker='o', color='red')
plt.xlabel('Year', color='g')
plt.ylabel('Count')
plt.title('Year Wise Employee Count')

plt.savefig('Line.png')
```
