# **More on Bar Plot in Seaborn**

Used for: Showing the mean (or Summary statistic (mean, median, etc.)) of a numerical value grouped by categories

Input: One categorical column + one numeric column

You can choose the statistic with **estimator=**

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv('employee_data.csv')
display(df.head(2))
```

```python
# total salary of each emp departmentwise
# categorical var: department
# numerical variable: salary
# estimate: total --= sum

plt.figure(figsize=(5, 3))
sns.barplot(x='department', y='salary', data=df, estimator='sum',
            palette='pastel', hue='gender')
plt.show()
```

```python
# avearge salary of each emp designation wise
# categorical var: designation
# numerical variable: salary
# estimate: average---mean

plt.figure(figsize=(5, 3))
sns.barplot(x='designation', y='salary', data=df, estimator='mean', palette='pastel')
plt.show()
```

```python
avg_sal = df.groupby('designation')['salary'].mean()
display(avg_sal)
```
